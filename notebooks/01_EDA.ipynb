```python
customer_df["TotalCharges"] = pd.to_numeric(customer_df["TotalCharges"],errors="coerce")
customer_df = customer_df.dropna()
```

```python
customer_df = customer_df.drop("customerID",axis=1)
```

To verify if customerID was removed

```python
customer_df.head()
```

One-hot encode all categorical features

```python
customer_df = pd.get_dummies(
    customer_df,
    columns=[
        "gender",
        "Partner",
        "Dependents",
        "PhoneService",
        "MultipleLines",
        "InternetService",
        "OnlineSecurity",
        "OnlineBackup",
        "DeviceProtection",
        "TechSupport",
        "StreamingTV",
        "StreamingMovies",
        "Contract",
        "PaperlessBilling",
        "PaymentMethod"
    ],
    dtype=int
)
```

```python
print(customer_df.columns.to_list())
```

A ML model also needs the target to be numeric,
Since there are only two possile values, we don't need one-hot encoding

```python
customer_df["Churn"] = customer_df["Churn"].map({"No":0,
                                                 "Yes":1})
```

```python
customer_df["Churn"].head()
```

Is there any str columns left? every column should now be numeric

```python
customer_df.dtypes
```

```python
x = customer_df.drop("Churn", axis = 1)
y = customer_df["Churn"]
```

```python
x.head()
```

```python
y.head()
```

```python
print(x.shape)
```

```python
print(y.shape)
```

Split the data, we use Scikit-Learn

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
```

Train the model

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import(accuracy_score, confusion_matrix,classification_report)
```

```python
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
confusionMatrix = confusion_matrix(y_test,y_pred)
print(confusionMatrix)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
scaled_model = LogisticRegression(max_iter = 1000)
```

```python
scaled_model.fit(x_train_scaled,y_train)
```

Make predictions

```python
scaled_y_pred = scaled_model.predict(x_test_scaled)
```

Evaluation

```python
scaled_accuracy = accuracy_score(y_test, scaled_y_pred)
print("Accuracy:", scaled_accuracy)
```

```python
scaled_confusionMatrix = confusion_matrix(y_test, scaled_y_pred)
print(scaled_confusionMatrix)
print(classification_report(y_test,scaled_y_pred))
```
